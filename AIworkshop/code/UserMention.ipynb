```python
def retrieveMention(df):
    df['new']=''
    
    reg=r"@.*? "
    for index, row in df.iterrows():
        df['new'][index]=';'.join(re.findall(reg, row['text']))
    return df
```

```python
def user_mention_df(dataset):
    userNameL = []
    mentionNameL = []
    lenOfmention = 0
    for index, row in dataset.iterrows():
        if len(dataset['new'].str.split(' ;', 100).tolist()[index]) == 1: # Â¶ÇÊûú‰∏Ä‰∏™Áî®Êà∑@Ë∂ÖËøá‰∏Ä‰∏™ÈÇ£‰πàÂ∞±Êää‰ªñ@ÁöÑÁî®Êà∑‰∏Ä‰∏™‰∏™ÈöîÁ¶ªÂØπÂ∫îÁî®Êà∑ÂàÜÂà´ÊîæÂú®‰∏ÄË°å
            userName = dataset['screen_name'].tolist()[index]
            mentionName = dataset['new'].tolist()[index] #Êú¨Êù•Â∞±@Èáå‰∏Ä‰∏™Áî®Êà∑ÔºåÂ∞±‰∏çÈúÄË¶ÅÈöîÁ¶ª‰∫ÜÔºå‰∏çÁÑ∂‰ºöÂá∫Áé∞list‰∏≠Êúâlist
            userNameL.append(userName)
            mentionNameL.append(mentionName)    
        else:
            for lenOfmention in range(len(dataset['new'].str.split(' ;', 100).tolist()[index])):
                userName = dataset['screen_name'].tolist()[index]
                mentionName = dataset['new'].str.split(' ;', 100).tolist()[index][lenOfmention]
                userNameL.append(userName)
                mentionNameL.append(mentionName)
    
    #List to dataframe
    c={"userName" : userNameL,
    "mentionName" : mentionNameL}
    
    data=pd.DataFrame(c)
    
    print('the lengh of user: ', len(userNameL))
    print('the lengh of mention:', len(mentionNameL))
    
    return data
```
